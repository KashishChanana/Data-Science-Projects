```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('hpt_small.csv')
data_prep= pd.get_dummies(data, drop_first= True)
```

```python
X= data_prep.iloc[:, :-1]
Y= data_prep.iloc[:, -1]
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier(random_state=1234)

from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(random_state=1234)

from sklearn.svm import SVC
svc= SVC(random_state=1234)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=1234)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rfc_param = {'n_estimators': [10,15,20], 'min_samples_split':[8,16], 'min_samples_leaf': [1,2,3,4,5]}
```

```python
rfc_grid= GridSearchCV(estimator= rfc, param_grid=rfc_param, scoring='accuracy', cv=10, return_train_score=True)
#3*2*5*10 = 300 jobs running here
```

```python
rfc_grid_fit=rfc_grid.fit(X, Y)

cv_result_rfc= rfc_grid_fit.cv_results_
```

```python
type(cv_result_rfc)
```

```python
cv_result_rfc= pd.DataFrame.from_dict(rfc_grid_fit.cv_results_)
```

```python
lr_param = {'C': [0.01, 0.1, 0.5, 1, 2, 5, 10], #inverse of lambda(regularisation parameter)
            'penalty': ['l2'],
           'solver': ['liblinear', 'lbfgs', 'saga']}
lr_grid= GridSearchCV(estimator= lr, param_grid=lr_param, scoring='accuracy', cv=10, return_train_score=True)

lr_grid_fit=lr_grid.fit(X, Y)

cv_result_lr= lr_grid_fit.cv_results_
```

```python
cv_result_lr= pd.DataFrame.from_dict(cv_result_lr)

cv_result_lr.sort_values(by=['rank_test_score'])
```

```python
svc_param= { 'C' : [0.01, 0.1, 0.5, 1, 2, 4 , 10],
            'kernel' : ['rbf', 'linear'],
            'gamma': [0.1, 0.25, 0.5, 1 , 5]
}
```

```python
svc_grid= GridSearchCV(estimator= svc, param_grid=svc_param, scoring='accuracy', cv=10, return_train_score=True)

svc_grid_fit=svc_grid.fit(X, Y)

cv_result_svc= svc_grid_fit.cv_results_
```

```python
cv_result_svc= pd.DataFrame.from_dict(cv_result_svc)

cv_result_svc.sort_values(by=['rank_test_score'])
```
