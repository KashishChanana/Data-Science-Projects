```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X= np.array([1,2,3,4])
Y= np.array([2,4,6,8])
# y=2x
```

```python
df= pd.DataFrame(X, columns=["X"])
```

```python
from sklearn.linear_model import Ridge
```

```python
def ridgeregressor(lmda):
    ridge= Ridge(alpha= lmda)
    ridge.fit(df,Y)
    ridge_coeff= ridge.coef_
    ridge_intercept= ridge.intercept_
    
    print("coefficient for "+ str(lmda)+ " : ", ridge_coeff)
    print("intercept for " + str(lmda)+ " : ", ridge_intercept)

    # y= 1.667*x+ 0.833 fr lambda =1.0
    x_plt= [0,1,2,3,4]
    y_plt= ridge.predict(pd.DataFrame(x_plt))
    
    plt.figure(1, figsize=(30,20))
    plt.plot(x_plt, y_plt)
    plt.ylim(ymin=0, ymax=9)
    plt.xlim(xmin=0, xmax=6)
    plt.text(x_plt[-1], y_plt[-1], 'y = '+ str('%.2f'%ridge_coeff) +'*x'+ ' + ' + str('%.2f'% ridge_intercept)+ ' for \u03BB or \u03B1 = '+ str(lmda), fontsize=12)
```

```python
lmda= [1.0, 0.0, 0.1, 10.0, 100.0]
for lm in lmda:
    ridgeregressor(lm)
```

```python
#high alpha -> decreases dependency on x
```

```python
from sklearn.linear_model import Lasso
```

```python
def lassoregressor(lmda):
    lasso= Lasso(alpha= lmda)
    lasso.fit(df,Y)
    lasso_coeff= lasso.coef_
    lasso_intercept= lasso.intercept_
    
    print("coefficient for "+ str(lmda)+ " : ", lasso_coeff)
    print("intercept for " + str(lmda)+ " : ", lasso_intercept)

    # y= 1.667*x+ 0.833 fr lambda =1.0
    x_plt= [0,1,2,3,4]
    y_plt= lasso.predict(pd.DataFrame(x_plt))
    
    plt.figure(2, figsize=(10,10))
    plt.plot(x_plt, y_plt)
    plt.ylim(ymin=0, ymax=9)
    plt.xlim(xmin=0, xmax=6)
    plt.text(x_plt[-1], y_plt[-1], 'y = '+ str('%.2f'%lasso_coeff) +'*x'+ ' + ' + str('%.2f'% lasso_intercept)+ ' for \u03BB or \u03B1 = '+ str(lmda), fontsize=12)
```

```python
lmda= [1.0, 0.0, 0.5, 1,2,4]
for lm in lmda:
    lassoregressor(lm)
```
